```python
import seaborn as sns
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x=make.values, y= make.index, palette='cubehelix')
plt.title('Top 10 Popular EV Makes')
plt.xlabel('Number of Vehicles registered')
plt.ylabel('Make')
plt.tight_layout()
plt.show()
```

```python
top_5_makes = make.head(3).index
top_makes = data[data['Make'].isin(top_5_makes)]
model_distri = top_makes.groupby(['Make','Model']).size().sort_values(ascending=False).reset_index(name='Number of Vehicles')
top_models = model_distri.head(10)
top_models
```

```python
plt.figure(figsize=(12,8))
sns.barplot(x='Number of Vehicles',y='Model',hue='Make',data=top_models,palette='viridis')
plt.title('Top Models in Top 3 Makes')
plt.xlabel('Number of Vehicles Registered')
plt.ylabel('Model')
plt.legend(title='Make',loc='center right')
plt.tight_layout()
plt.show()
```

```python
ev_type = data['Electric Vehicle Type'].value_counts()
ev_type
```
