```python
df = df.drop(columns='Cabin')
df
```

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df
```

```python
df['Age'] = df['Age'].astype(int)

df
```

```python
import re
```

```python
def split_name(name):
    name = re.sub(r'\(.*?\)', '', name).strip()
    parts = name.split(',')
    Nickname = parts[0].strip()
    TitleName = parts[1].strip()
    return pd.Series([Nickname, TitleName])
df[['Nickname', 'TitleName']] = df['Name'].apply(split_name)
df
```

```python
df.drop(columns='Name', inplace=True)
```

```python
df
```

```python
df['Fare'] = df['Fare'].round().astype(int)
df
```

```python
df2 = df.set_index(['PassengerId', 'Embarked'])
```

```python
df2
```

```python
df2.filter(like='S', axis=0)
```

```python
for x in df2.index:
    if df2.loc[x, 'Survived'] == 0:
        df2.drop(x, inplace=True)
df2
```

```python
df2[df2['Age']>= 30].sort_values(by=['Sex','Embarked','Survived'], ascending=[True,False,True])
```

```python
df2['Ticket'] = df2['Ticket'].str.replace(' ', '').str.upper()
df2['Ticket'].unique()
```

```python
def extract_ticket(ticket):
    match = re.match(r'([A-Za-z./]+)?(\d+)', ticket)
    if match:
        prefix = match.group(1) if match.group(1) else ''
        number = match.group(2)
        return pd.Series([prefix, number])
    else:
        return pd.Series([ticket,''])
df2[['Ticketpref','TicketNum']] = df2['Ticket'].apply(extract_ticket)
df2
```

```python
df2.drop(columns='Ticket', inplace=True)
```
